```python
import random
WORDLIST = 'wordlist.txt'
def get_random_word(min_word_length):
    """Get a random word from the wordlist using no extra memory."""
    num_words_processed = 0
    curr_word = None
    with open(WORDLIST, 'r') as f:
        for word in f:
            if '(' in word or ')' in word:
                continue
            word = word.strip().lower()
            if len(word) < min_word_length:
                continue
            num_words_processed += 1
            if random.randint(1, num_words_processed) == 1:
                curr_word = word
    return curr_word
print('Starting a game of Hangman...')
name=input('Enter your name: ')
print(name,"Welcome to the game of Hangman, Good Luck ! ")
mi=int(input("Enter the mininum word length do you want? [4-10]: "))
print('Selecting a word...')
word1=get_random_word(mi).upper()
word=list(word1)
print('Hint:')
print('The First word is',word[0],'last word is',word[-1])
gword=list("_"*len(word))
gword[0],gword[-1]=word[0],word[-1]
word[0],word[-1]='_','_'
print(' '.join(gword))
print('You have',int(2.5*len(word)),'attempts')
ip=input("Enter the guess letter ")
c=int(2.5*len(word))
while(True):
    if(ip.upper() in word):
        c-=1
        index=word.index(ip.upper())
        gword[index]=ip.upper()
        print(' '.join(gword))
        word[index]="_"
    else:
        print(ip,'is Not in the word')    
    if "_" not in gword:
        print(" You Won ")
        break
    else:
        ip=input("Enter the another guess letter ")
        c-=1
        print(c," Attempts left")
    if(c<=0):
        print("The number of avilable attempts finished")
        print('The word is ',word1)
        print("Better luck next Time")
        break
```
